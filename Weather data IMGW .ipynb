```python
import pandas as pd
from io import BytesIO
import urllib.request
from zipfile import ZipFile
from datetime import date
```

```python
def m_zero(i):
    return f'0{str(i)}' if i<10 else str(i)
```

## --testing loop for all files--

```python
rok = date.today().year
data = [['Kod stacji','Nazwa stacji', 'Rok', 'Miesiac','Dzien', 'Maksymalna temperatura dobowa (°C)', 'Status pomiaru TMAX', 'Minimalna temperatura dobowa (°C)', 'Status pomiaru TMIN', 'Srednia temperatura dobowa (°C)', 'Status pomiaru STD', 'Temperatura minimalna przy gruncie (°C)', 'Status pomiaru TMNG', 'Suma dobowa opadow (mm)', 'Status pomiaru SMDB','Rodzaj opadu  (S/W/ )', 'Wysokosc pokrywy snieznej (cm)', 'Status pomiaru PKSN' ],]

for i in range (2022,rok+1):
    for j in range (1, 13):
        url = f'https://danepubliczne.imgw.pl/data/dane_pomiarowo_obserwacyjne/dane_meteorologiczne/dobowe/klimat/{str(i)}/{str(i)}_{m_zero(j)}_k.zip'
        req = urllib.request.Request(url)
        try:
            response = urllib.request.urlopen(req)
            myzip = ZipFile(BytesIO(response.read()))
            for line in myzip.open(myzip.namelist()[0]):
                data.append(line.decode('utf-8', errors='ignore').replace('"','').split(','))
            print(f'Successfully processed {str(i)}_{m_zero(j)}_k.zip')
        except urllib.error.HTTPError as e:
            print(f'{str(i)}_{m_zero(j)}_k.zip - {e.code} - Error while trying to access file')
       
        
        
```

```python
df = pd.DataFrame(data)
df.columns = df.iloc[0]
df = df.drop(df.index[0])

df = df.astype({'Rok':'int64', "Miesiac":'int64','Dzien':'int64', 'Maksymalna temperatura dobowa (°C)':'float64', 'Minimalna temperatura dobowa (°C)':'float64', 'Srednia temperatura dobowa (°C)':'float64', 'Temperatura minimalna przy gruncie (°C)':'float64',  'Suma dobowa opadow (mm)':'float64', "Wysokosc pokrywy snieznej (cm)":'int64'})

df.dtypes
```
